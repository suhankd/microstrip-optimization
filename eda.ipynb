```python
import pandas as pd
import numpy as np

import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.DataFrame()

for i in range(1, 6):

    _ = pd.read_csv(f"Data\\generatedValues_{i}.csv")
    df = df._append(_)

df.describe()
```

```python
sns.histplot(df['area'], kde=True)
plt.title("Distribution of Area")
plt.show()

df['log_area'] = np.log1p(df['area'])
sns.histplot(df['log_area'], kde=True)
plt.title("Distribution of log(Area)")
plt.show()
```

Here, as the distribution is skewed towards the left, we apply the log transform to better highlight data features.

```python
corr = (df.drop("area", axis = 1)).corr()

plt.figure(figsize=(10,10))
sns.heatmap(corr, cmap = "coolwarm")
plt.title("Area Correlation Heatmap")
plt.show()
```
